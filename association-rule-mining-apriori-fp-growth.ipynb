  </p>

  <p style="color:#555555; font-size:16px;">
      A typical rule looks like this:
      <span style="background-color:#E67E22; color:white; padding:2px 6px; border-radius:4px;">{Milk, Bread} â†’ {Butter}</span>

  <!-- Apriori Algorithm -->
  <h2 style="color:#E67E22; font-size:28px; margin-bottom:15px;">ğŸ”¹ Apriori Algorithm</h2>
  <p>
    The <b>Apriori Algorithm</b> is a classic method for <b>frequent itemset mining</b> and generating <b>association rules</b>.
  </p>

  <h3 style="color:#2E86C1;">How Apriori Works:</h3>
  <ol>
    <li><b>Frequent Itemsets Generation</b>  
      - Starts with single items (length = 1).  
      - Iteratively extends to larger itemsets using the <b>Apriori property</b>:<br>
      <span style="font-style:italic; color:#E67E22;">"All subsets of a frequent itemset must also be frequent." âœ…</span>
    </li>
    <li><b>Candidate Pruning</b>  
      - Itemsets that do not meet minimum support are discarded.  
      - Reduces the search space significantly.
    </li>
    <li><b>Rule Generation</b>  
      - Once frequent itemsets are found, rules are generated.  
      - Each rule is evaluated with <b>confidence</b> and <b>lift</b> metrics.
    </li>
  </ol>

  <h3 style="color:#2E86C1;">Advantages of Apriori</h3>
  <ul>
    <li>Easy to understand and implement.</li>
    <li>Works well for <b>small datasets</b>.</li>
  </ul>

  <h3 style="color:#2E86C1;">Disadvantages</h3>
  <ul>
    <li>Generates <b>many candidate sets</b>, especially for large datasets.</li>
    <li>Computationally expensive. âš ï¸</li>
  </ul>

  <hr style="border: 1px solid #2E86C1; margin:25px 0;">

  <!-- FP-Growth -->
  <h2 style="color:#E67E22; font-size:28px; margin-bottom:15px;">ğŸ”¹ FP-Growth Algorithm</h2>
  <p>
    The <b>Frequent Pattern Growth (FP-Growth) Algorithm</b> is an <b>optimized alternative</b> to Apriori.
  </p>

  <h3 style="color:#2E86C1;">How FP-Growth Works:</h3>
  <ol>
    <li><b>FP-Tree Construction</b>  
      - Converts the dataset into a <b>compact prefix-tree structure (FP-Tree)</b>.  
      - Common itemsets share the same path â†’ reduces redundancy.
    </li>
    <li><b>Frequent Pattern Mining</b>  
      - Recursively mines the FP-Tree to extract <b>frequent itemsets</b> without candidate generation.  
      - Uses <b>conditional pattern bases</b> and <b>conditional FP-Trees</b>.
    </li>
  </ol>

  <h3 style="color:#2E86C1;">Advantages of FP-Growth</h3>
  <ul>
    <li>Much faster than Apriori for <b>large datasets</b>. ğŸš€</li>
    <li>No need for generating candidate sets.</li>
    <li>Efficient memory usage with FP-Tree.</li>
  </ul>

  <h3 style="color:#2E86C1;">Disadvantages</h3>
  <ul>
    <li>Slightly more complex to implement.</li>
    <li>FP-Tree construction may be expensive for extremely large datasets with long transactions.</li>
  </ul>

  <hr style="border: 1px solid #2E86C1; margin:25px 0;">

  <li>âœ” Forms the foundation for accurate frequent itemset mining</li>
</ul>

<blockquote>
At this stage, the dataset is <strong>perfectly prepared</strong> for applying <strong>Apriori</strong> and <strong>FP-Growth algorithms</strong>.

<h2 style="color:#2E86C1; font-size:28px; margin-bottom:15px;">ğŸ“Š Frequent Itemsets & Association Rule Mining (Apriori)</h2>

<p>In this step, we apply the <strong>Apriori algorithm</strong> on the binary basket data to uncover <strong>frequent itemsets</strong> and transform them into <strong>association rules</strong> that reveal customer purchasing behavior.</p>

<hr>

<h3 style="color:#2E86C1; font-size:14px; margin-bottom:15px;">ğŸ”¹ Mining Frequent Itemsets</h3>

<p>The Apriori algorithm is executed with a <strong>minimum support threshold of 0.05</strong>, meaning:</p>
<ul>
  <li>âœ” An itemset must appear in <strong>at least 5% of all transactions</strong></li>
  <li>âœ” Helps filter out rare or noisy combinations</li>
  <li>âœ” Ensures statistically meaningful patterns</li>
</ul>

<p>After mining, the frequent itemsets are <strong>sorted in descending order of support</strong>, and the <strong>Top 20 most frequent itemsets</strong> are displayed.</p>

<p>ğŸ“Œ These itemsets represent:</p>

<h3 style="color:#2E86C1; font-size:14px; margin-bottom:15px;">ğŸ”¹ Generating Association Rules</h3>

<p>Once frequent itemsets are identified, they are converted into <strong>association rules</strong> of the form:</p>
<p><strong>Antecedent â†’ Consequent</strong></p>

<p>A <strong>confidence threshold of 0.2</strong> is applied:</p>
<ul>
  <li>âœ” Captures realistic and interpretable rules</li>

  <li>âœ” Suitable for retail and grocery datasets</li>
</ul>

<p>Each rule is evaluated using three key metrics:</p>
<ul>
  <li><strong>Support</strong> â†’ Overall frequency of the rule</li>
  <li><strong>Confidence</strong> â†’ Likelihood of purchasing the consequent given the antecedent</li>
  <li><strong>Lift</strong> â†’ Strength of association compared to random chance</li>
</ul>

<hr>

<h3 style="color:#2E86C1; font-size:14px; margin-bottom:15px;">ğŸ”¹ Sorting Rules by Lift (Interpretability First)</h3>

  <li><strong>Lift â‰ˆ 1</strong> â†’ Items are independent âš ï¸</li>
  <li><strong>Lift &lt; 1</strong> â†’ Negative association âŒ</li>

<h2 style="color:#2E86C1; font-size:28px; margin-bottom:15px;">ğŸ”¹ FP-Growth Algorithm Implementation</h2>

<p>The <strong>Frequent Pattern Growth (FP-Growth) Algorithm</strong> is a faster and more memory-efficient alternative to the Apriori algorithm.</p>

<p>Unlike Apriori, FP-Growth <strong>does not generate candidate itemsets</strong>. Instead, it builds a compact <strong>FP-Tree (Frequent Pattern Tree)</strong> and mines frequent patterns recursively.</p>

<h3 style="color:#2E86C1; font-size:14px; margin-bottom:15px;">Key Advantages:</h3>
<ul>
  <li>ğŸš€ Much faster on <strong>large datasets</strong></li>
  <li>ğŸ’¾ Efficient memory usage</li>
  <li>âŒ No candidate generation required</li>
</ul>

<h3 style="color:#2E86C1; font-size:14px; margin-bottom:15px;">How FP-Growth Works:</h3>
<ol>
  <li><strong>FP-Tree Construction</strong>
    <ul>
      <li>Converts transactions into a <strong>prefix-tree structure</strong></li>
      <li>Common paths are shared â†’ reduces redundancy</li>
    </ul>
  </li>
  <li><strong>Mining Frequent Patterns</strong>
    <ul>
      <li>Recursively extracts <strong>frequent itemsets</strong> from FP-Tree</li>
      <li>Uses <strong>conditional pattern bases</strong> for efficiency</li>
    </ul>
  </li>
</ol>

<blockquote>
In the next code cell, we will implement FP-Growth on the same transaction dataset and generate frequent itemsets & association rules. ğŸ›’ğŸ“Š
</blockquote>

</div>


```python
# Include all itemsets, even length 1
frequent_itemsets_fp_all = fpgrowth(basket, min_support=0.05, use_colnames=True)

# Keep itemsets of length >=2 for display, but keep original for rules
frequent_itemsets_fp = frequent_itemsets_fp_all[frequent_itemsets_fp_all['itemsets'].apply(lambda x: len(x) >= 2)].copy()

print("Top 20 frequent itemsets (length >=2):")
print(frequent_itemsets_fp.sort_values(by='support', ascending=False).head(20))
```
