# Exploratory Data Analysis

```python
df_sorted_by_subscribers = df.sort_values(by='Subscribers (more then)', ascending=False)
df_sorted_by_views = df.sort_values(by='Views (more then)', ascending=False)
df_sorted_by_videos = df.sort_values(by='Videos (more then)', ascending=False)
df_sorted_by_last_30_subscribers = df.sort_values(by='Last 30 Days Subscribers', ascending=False)
df_sorted_by_last_30_views = df.sort_values(by='Last 30 Days Views', ascending=False)
df_sorted_by_most_popular_video = df.sort_values(by='Max Views on Video', ascending=False)
```

### Top 10 Channels by Subscriber Count

```python
top_10_channels_by_subscribers = df_sorted_by_subscribers[['Chennel Name', 'Subscribers (more then)']].head(10)
```

```python
top_10_channels_by_subscribers
```

```python
plt.figure(figsize=(10,6))

plt.barh(top_10_channels_by_subscribers['Chennel Name'], top_10_channels_by_subscribers['Subscribers (more then)'], color='green')
plt.xlabel('Subscribers (more than)', fontsize=12)
plt.ylabel('Channel Name', fontsize=12)
plt.title('Top 10 YouTube Channels by Subscribers', fontsize=14)

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

### Correlation Matrix of Dataset

```python
# Select only the numeric columns
numeric_columns = ['Subscribers (more then)', 'Views (more then)', 'Videos (more then)', 'Last 30 Days Views', 'Last 30 Days Subscribers', 'Max Views on Video']

# Calculate the correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Generate a heatmap to visualize the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")

# Add title
plt.title('Correlation Matrix of Numeric Features', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```
