# **üìä Overall Trends**

#### **üìä 1Ô∏è‚É£ Total Wheat Production Over Time (Africa)**
**‚ùì Question**

How has total wheat production changed over time across Africa?

**üìå Idea**

Aggregate production across all countries per year and plot a time series.

**üìà Plot Type**

Line plot (Year vs Total Production)

**üß† Why**

This shows the macro trend clearly ‚Äî the backbone of your EDA.

```python
# Aggregate total production per year
africa_total = (
    df.groupby("Year")["Value"]
      .sum()
      .reset_index()
)

plt.figure(figsize=(12, 6))
plt.plot(africa_total["Year"], africa_total["Value"])
plt.title("Total Wheat Production in Africa (1961‚Äì2025)")
plt.xlabel("Year")
plt.ylabel("Total Wheat Production")
plt.grid(True)
plt.show()
```

#### **üìà 2Ô∏è‚É£ Trend Direction (Increasing, Stagnating, or Declining)**
**‚ùì Question**

Is overall production increasing, stagnating, or declining?

**üìå Idea**

Add a trend line (rolling mean) to smooth short-term noise.

**üìà Plot Type**

Line plot + rolling average

**üß† Why**

Rolling mean reveals the true direction of the time series.

```python
# Rolling mean (5-year window)
africa_total["Rolling_5Y"] = africa_total["Value"].rolling(window=5).mean()

plt.figure(figsize=(12, 6))
plt.plot(africa_total["Year"], africa_total["Value"], label="Annual Production")
plt.plot(africa_total["Year"], africa_total["Rolling_5Y"], linewidth=3, label="5-Year Rolling Average")

plt.title("Africa Wheat Production Trend with Rolling Average")
plt.xlabel("Year")
plt.ylabel("Total Wheat Production")
plt.legend()
plt.grid(True)
plt.show()
```

#### **üìâ 3Ô∏è‚É£ Structural Shifts & Breakpoints**
**‚ùì Question**

Are there major structural shifts or breakpoints in production trends?

**üìå Idea**

Plot Year-over-Year percentage change of total production.

**üìà Plot Type**

Line plot of YoY % change

**üß† Why**

**Sudden spikes or drops often indicate:**

- Climate shocks

- Policy changes

- Technological adoption

- Data regime shifts

```python
# Compute YoY % change of total production
africa_total["YoY_Change_Total"] = africa_total["Value"].pct_change() * 100

plt.figure(figsize=(12, 6))
plt.plot(africa_total["Year"], africa_total["YoY_Change_Total"])
plt.axhline(0, linestyle="--")

plt.title("Year-over-Year Change in Africa's Total Wheat Production")
plt.xlabel("Year")
plt.ylabel("YoY Change (%)")
plt.grid(True)
plt.show()
```

# **üåç Country-Level Insights**

#### **üåç 1Ô∏è‚É£ Top Wheat Producers Overall**
**‚ùì Question**

Which countries are the top wheat producers overall?

**üìå Idea**

Aggregate total production across all years per country and rank.

**üìà Plot Type**

Horizontal bar chart (best for rankings)

```python
country_total = (
    df.groupby("Country")["Value"]
      .sum()
      .sort_values()
)

plt.figure(figsize=(10, 6))
plt.barh(country_total.index, country_total.values)
plt.title("Total Wheat Production by Country (1961‚Äì2025)")
plt.xlabel("Total Wheat Production")
plt.ylabel("Country")
plt.grid(axis="x")
plt.show()
```

#### **üìà 2Ô∏è‚É£ Production Trends by Country (Time Series)**
**‚ùì Question**

How do production trends differ by country?

**üìå Idea**

Plot one line per country over time.

**üìà Plot Type**

Multi-line time series

```python
plt.figure(figsize=(13, 7))

for country in df["Country"].unique():
    country_data = df[df["Country"] == country]
    plt.plot(country_data["Year"], country_data["Value"], label=country)

plt.title("Wheat Production Trends by Country")
plt.xlabel("Year")
plt.ylabel("Wheat Production")
plt.legend()
plt.grid(True)
plt.show()
```

#### **üìä 3Ô∏è‚É£ Consistent Long-Term Growth**
**‚ùì Question**

Which countries show consistent long-term growth?

**üìå Idea**

Smooth each country‚Äôs production using a rolling mean.

**üìà Plot Type**

Smoothed trend lines

```python
plt.figure(figsize=(13, 7))

for country in df["Country"].unique():
    country_data = df[df["Country"] == country].copy()
    country_data["Rolling_5Y"] = country_data["Value"].rolling(5).mean()
    plt.plot(country_data["Year"], country_data["Rolling_5Y"], label=country)

plt.title("Smoothed Wheat Production Trends (5-Year Rolling Average)")
plt.xlabel("Year")
plt.ylabel("Wheat Production (Smoothed)")
plt.legend()
plt.grid(True)
plt.show()
```

#### **üìâ 4Ô∏è‚É£ Volatility & Instability by Country**
**‚ùì Question**

Which countries show high volatility or instability?

**üìå Idea**

Measure standard deviation of YoY Change per country.

**üìà Plot Type**

Bar chart (volatility ranking)

```python
volatility = (
    df.groupby("Country")["YoY Change"]
      .std()
      .sort_values()
)

plt.figure(figsize=(10, 6))
plt.barh(volatility.index, volatility.values)
plt.title("Production Volatility by Country (YoY Std Dev)")
plt.xlabel("Volatility (Std Dev of YoY Change)")
plt.ylabel("Country")
plt.grid(axis="x")
plt.show()
```

# **üìà Growth & YoY Analysis**

#### **üìà 1Ô∏è‚É£ Average YoY Growth Rate by Country**
**‚ùì Question**

What is the average YoY growth rate for each country?

**üìå Idea**

Compute mean YoY Change per country and rank them.

**üìä Plot Type**

Horizontal bar chart (ranking)

```python
avg_yoy = (
    df.groupby("Country")["YoY Change"]
      .mean()
      .sort_values()
)

plt.figure(figsize=(10, 6))
plt.barh(avg_yoy.index, avg_yoy.values)
plt.title("Average Year-over-Year Wheat Production Growth by Country")
plt.xlabel("Average YoY Growth (%)")
plt.ylabel("Country")
plt.axvline(0, linestyle="--")
plt.grid(axis="x")
plt.show()
```

#### **üöÄ 2Ô∏è‚É£ Largest YoY Increase (Positive Shock)**
**‚ùì Question**

Which country experienced the largest YoY increase?

**üìå Idea**

Find maximum YoY Change per country.

**üìä Plot Type**

Bar chart (extreme growth events)

```python
max_yoy = (
    df.groupby("Country")["YoY Change"]
      .max()
      .sort_values()
)

plt.figure(figsize=(10, 6))
plt.barh(max_yoy.index, max_yoy.values)
plt.title("Largest Year-over-Year Increase in Wheat Production")
plt.xlabel("Maximum YoY Increase (%)")
plt.ylabel("Country")
plt.grid(axis="x")
plt.show()
```

#### **üìâ 3Ô∏è‚É£ Largest YoY Decline (Negative Shock)**
**‚ùì Question**

Which country experienced the largest YoY decline?

**üìå Idea**

Find minimum YoY Change per country.

**üìä Plot Type**

Bar chart (negative extremes)

```python
min_yoy = (
    df.groupby("Country")["YoY Change"]
      .min()
      .sort_values()
)

plt.figure(figsize=(10, 6))
plt.barh(min_yoy.index, min_yoy.values)
plt.title("Largest Year-over-Year Decline in Wheat Production")
plt.xlabel("Minimum YoY Change (%)")
plt.ylabel("Country")
plt.grid(axis="x")
plt.show()
```

#### **üìä 4Ô∏è‚É£ Are YoY Changes Becoming More Volatile Over Time?**
**‚ùì Question**

Are YoY changes becoming more volatile in recent years?

**üìå Idea**

Measure rolling standard deviation of YoY change across all countries.

**üìà Plot Type**

Line plot (volatility trend)

```python
# Aggregate YoY change across all countries per year
yoy_yearly = (
    df.groupby("Year")["YoY Change"]
      .mean()
      .reset_index()
)

# Rolling volatility (5-year window)
yoy_yearly["Rolling_Volatility"] = yoy_yearly["YoY Change"].rolling(5).std()

plt.figure(figsize=(12, 6))
plt.plot(yoy_yearly["Year"], yoy_yearly["Rolling_Volatility"])
plt.title("Rolling Volatility of Wheat Production YoY Change (Africa)")
plt.xlabel("Year")
plt.ylabel("Rolling Std Dev of YoY Change")
plt.grid(True)
plt.show()
```

# **üèÜ Comparison & Ranking**

#### **üèÜ 1Ô∏è‚É£ Country Rankings by Production Over Time**
**‚ùì Question**

How do countries rank by production over time?

**üìå Idea**

Rank countries within each year based on production and track rank changes.

**üìà Plot Type**

Line plot of ranks (lower = better)

```python
# Rank countries by production for each year
rank_df = df.copy()
rank_df["Rank"] = rank_df.groupby("Year")["Value"].rank(
    ascending=False, method="min"
)

plt.figure(figsize=(13, 7))

for country in rank_df["Country"].unique():
    cdata = rank_df[rank_df["Country"] == country]
    plt.plot(cdata["Year"], cdata["Rank"], label=country)

plt.gca().invert_yaxis()  # Rank 1 at top
plt.title("Country Rankings by Wheat Production Over Time")
plt.xlabel("Year")
plt.ylabel("Rank (1 = Highest Producer)")
plt.legend()
plt.grid(True)
plt.show()
```

#### **üöÄ 2Ô∏è‚É£ Catching Up to Traditional Leaders**
**‚ùì Question**

Which countries are catching up to traditional leaders?

**üìå Idea**

Compare early vs recent average production.

**üìà Plot Type**

Slope-style comparison using bars

```python
early_period = df[df["Year"] <= 1980].groupby("Country")["Value"].mean()
recent_period = df[df["Year"] >= 2010].groupby("Country")["Value"].mean()

growth_ratio = ((recent_period - early_period) / early_period * 100).sort_values()

plt.figure(figsize=(10, 6))
plt.barh(growth_ratio.index, growth_ratio.values)
plt.title("Production Growth: Recent Years vs Early Period")
plt.xlabel("Growth Since Early Period (%)")
plt.ylabel("Country")
plt.axvline(0, linestyle="--")
plt.grid(axis="x")
plt.show()
```

# **üß† Reliability & Risk Signals**

#### **üß† 1Ô∏è‚É£ Proportion of Estimated vs Official Data**
**‚ùì Question**

What proportion of data is Estimated vs Official?

**üìå Idea**

Count the number of records by Flag.

**üìà Plot Type**

Pie chart or bar chart

```python
flag_counts = df["Flag"].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(flag_counts.values, labels=flag_counts.index, autopct="%1.1f%%", startangle=140, colors=["#66b3ff","#ff9999"])
plt.title("Proportion of Estimated vs Official Data")
plt.show()
```

#### **üìä 2Ô∏è‚É£ Estimated vs Official Behavior**
**‚ùì Question**

Do Estimated values show different behavior than Official ones?

**üìå Idea**

Compare Value distributions of Estimated vs Official using boxplots.

**üìà Plot Type**

Boxplot per Flag

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Flag", y="Value", data=df, palette=["#66b3ff","#ff9999"])
plt.title("Wheat Production: Estimated vs Official Records")
plt.ylabel("Wheat Production")
plt.xlabel("Record Type")
plt.show()
```

#### **üìâ 3Ô∏è‚É£ Most Resilient Countries to Production Shocks**
**‚ùì Question**

Which countries appear most resilient to production shocks?

**üìå Idea**

Measure volatility using YoY Change standard deviation (lower = more resilient).

**üìà Plot Type**

Horizontal bar chart

```python
resilience = df.groupby("Country")["YoY Change"].std().sort_values()

plt.figure(figsize=(10, 6))
plt.barh(resilience.index, resilience.values, color="#66b3ff")
plt.title("Resilience of Countries to Wheat Production Shocks (Lower Std Dev = More Resilient)")
plt.xlabel("Volatility (Std Dev of YoY Change)")
plt.ylabel("Country")
plt.grid(axis="x")
plt.show()
```

#### **üìà 4Ô∏è‚É£ Most Vulnerable Countries to Sharp Declines**
**‚ùì Question**

Which countries appear most vulnerable to sharp declines?

**üìå Idea**

Measure minimum YoY Change per country (most negative = most vulnerable).

**üìà Plot Type**

Horizontal bar chart

```python
vulnerability = df.groupby("Country")["YoY Change"].min().sort_values()

plt.figure(figsize=(10, 6))
plt.barh(vulnerability.index, vulnerability.values, color="#ff6666")
plt.title("Vulnerability of Countries to Sharp Wheat Production Declines")
plt.xlabel("Largest Negative YoY Change (%)")
plt.ylabel("Country")
plt.grid(axis="x")
plt.show()
```
