```python
import os
from captcha.image import ImageCaptcha  
import random  

# Create a folder to store CAPTCHA images
os.makedirs("captchas", exist_ok=True)

# Create a CAPTCHA generator  
image = ImageCaptcha(width=280, height=90, fonts=['arial.ttf', 'comic.ttf'], font_sizes=[40, 50, 60])  


# Number of images to generate
num_images = 50000  

for i in range(num_images):
    # Generate random text (6-character alphanumeric)
    text = ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=6))
    
    # Generate and save the image
    image_path = f"captchas/{text}.png"
    image.write(text, image_path)
    
    # Print progress
    if i % 500 == 0:
        print(f"Generated {i} images...")

print("Dataset generation complete!")

```

```python
import pyttsx3
import random
import os

os.makedirs("audio_captchas", exist_ok=True)
engine = pyttsx3.init()

for i in range(50000):
    text = ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=6))
    audio_path = f"audio_captchas/{text}.mp3"
    
    engine.save_to_file(text, audio_path)
    engine.runAndWait()

    if i % 500 == 0:
        print(f"Generated {i} audio CAPTCHAs...")

print("Offline Audio CAPTCHA dataset created!")

```
