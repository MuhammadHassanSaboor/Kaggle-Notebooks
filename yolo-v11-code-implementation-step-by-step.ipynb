# Importing Library

```python
from ultralytics import YOLO
```

# Loading a Model

```python
model = YOLO("yolo11n.pt")
```

# Training of Model

Please use GPU to make it fast

```python
train_results = model.train(

    data = "coco8.yaml",
    epochs = 100,
    imgsz=640
)
```

# Evaluate Model on Validation Set

```python
metrics = model.val()
```

# Perform Object Detection on an Image

```python
results = model("/kaggle/working/datasets/coco8/images/train/000000000034.jpg")
results[0].show()
```
