# **üåç World Famous Places: Exploratory Data Analysis (EDA)** 

Welcome to this EDA notebook on **World Famous Places**!  
This dataset contains detailed information about popular tourist destinations across the globe, including their geographical location, visitor statistics, historical background, cultural significance, entry fees, and tourism revenue.  

The goal of this analysis is to explore global tourism trends, identify patterns in visitor behavior, understand economic impact, and uncover insights that can help travelers, researchers, and tourism planners.  

This notebook is organized into **11 key sections**, each focusing on a different aspect of tourism analysis:  

1. **üåç Global Tourism Popularity** ‚Äì Discover the most and least visited places, compare popularity across countries and regions, and identify cities with the highest tourist footfall.  

2. **üèõÔ∏è Place Type Analysis** ‚Äì Analyze the types of attractions (Monuments, Natural Wonders, Religious Sites, Historical Sites), compare visitor numbers, and explore revenue generation by place type.  

3. **üèÜ UNESCO World Heritage Impact** ‚Äì Compare UNESCO vs non-UNESCO sites in terms of visitors and revenue, and study regional distributions of UNESCO sites.  

4. **üí∞ Tourism Revenue Insights** ‚Äì Identify top revenue-generating destinations, analyze the relationship between visitors and revenue, explore entry fee impact, and detect high-value locations.  

5. **üéüÔ∏è Entry Fee Patterns** ‚Äì Compare free and paid attractions, study how entry fees affect visitors, and identify destinations that remain popular despite high fees.  

6. **üï∞Ô∏è Historical & Cultural Trends** ‚Äì Explore the age of attractions, compare ancient vs modern sites, and identify regions rich in historical landmarks.  

7. **‚è±Ô∏è Visitor Behavior Analysis** ‚Äì Analyze average visit durations, identify destinations where tourists spend more time, and understand places offering deeper engagement experiences.  

8. **üå§Ô∏è Best Time to Visit** ‚Äì Identify optimal visiting months, detect seasonal tourism patterns, and study the influence of climate and festivals on tourism.  

9. **üåé Regional Comparison** ‚Äì Compare tourism trends across continents, identify regions with the highest visitors, revenue, and concentration of famous landmarks.  

10. **üìå Signature Attractions** ‚Äì Understand what each place is most famous for, discover cultural, natural, or architectural themes, and explore regional specialization.  

11. **üìä Correlation & Relationships** ‚Äì Study simple correlations between visitors, revenue, entry fees, and visit duration to understand underlying patterns in tourism.  

This analysis is designed to be **beginner-friendly**, visually engaging, and insightful, helping both data enthusiasts and tourism professionals gain a deeper understanding of **global travel patterns and tourism dynamics**.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/world-famous-places/world_famous_places_2024.csv")
```

```python
df.info()
```

# **üåç 1. Global Tourism Popularity**

### **üìä 1Ô∏è‚É£ Most & Least Visited Places (Top & Bottom)**
**üîπ Purpose**

- Identify top tourist attractions
- Highlight least visited famous places

**üìà Plot: Bar Chart (Sorted)**

```python
df_sorted = df.sort_values(by="Annual_Visitors_Millions", ascending=False)

plt.figure(figsize=(12,6))
plt.bar(df_sorted["Place_Name"], df_sorted["Annual_Visitors_Millions"])
plt.xticks(rotation=75)
plt.xlabel("Famous Places")
plt.ylabel("Annual Visitors (Millions)")
plt.title("Most and Least Visited Famous Places Worldwide")
plt.tight_layout()
plt.show()
```

### **üåç 2Ô∏è‚É£ Tourism Popularity Across Countries**
**üîπ Purpose**

- Compare total tourism popularity by country
- Identify top tourism countries

**üìà Plot: Country-wise Visitors**

```python
country_visitors = df.groupby("Country")["Annual_Visitors_Millions"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.bar(country_visitors.index, country_visitors.values)
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Total Annual Visitors (Millions)")
plt.title("Tourism Popularity Across Countries")
plt.tight_layout()
plt.show()
```

### **üåé 3Ô∏è‚É£ Tourism Popularity by Region**
**üîπ Purpose**

- Understand regional tourism strength
- Compare continents/regions easily

**üìä Plot: Pie Chart (Beginner Friendly)**

```python
region_visitors = df.groupby("Region")["Annual_Visitors_Millions"].sum()

plt.figure(figsize=(7,7))
plt.pie(region_visitors, labels=region_visitors.index, autopct='%1.1f%%', startangle=140)
plt.title("Tourism Share by Region")
plt.show()
```

### **üèôÔ∏è 4Ô∏è‚É£ Cities with Highest Tourist Footfall**
**üîπ Purpose**

- Discover most popular tourist cities
- Useful for travel analytics & recommendations

**üìà Plot: City-wise Visitors**

```python
city_visitors = df.groupby("City")["Annual_Visitors_Millions"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
plt.bar(city_visitors.index, city_visitors.values)
plt.xticks(rotation=75)
plt.xlabel("City")
plt.ylabel("Annual Visitors (Millions)")
plt.title("Cities with Highest Tourist Footfall")
plt.tight_layout()
plt.show()
```

# **üï∞Ô∏è 6. Historical & Cultural Trends**

### **üï∞Ô∏è 1Ô∏è‚É£ Age Distribution of Famous Places (Year Built)**
**üîπ Purpose**

- Understand how old or modern famous places are
- Visualize historical spread

**üìä Plot: Bar Chart (Sorted by Year Built)**

```python
# Convert Year_Built to numeric (errors ignored if text exists)
df["Year_Built_Num"] = pd.to_numeric(df["Year_Built"], errors="coerce")

df_year_sorted = df.sort_values("Year_Built_Num")

plt.figure(figsize=(12,6))
plt.bar(df_year_sorted["Place_Name"], df_year_sorted["Year_Built_Num"])
plt.xticks(rotation=75)
plt.xlabel("Famous Places")
plt.ylabel("Year Built")
plt.title("Historical Timeline of Famous Places")
plt.tight_layout()
plt.show()
```

### **üèõÔ∏è 2Ô∏è‚É£ Ancient vs Modern Attractions (Visitor Comparison)**
**üîπ Purpose**

- Compare tourist interest in old vs new places
- Beginner-friendly categorization

**üìå Definition (Simple & Explainable)**

- Ancient ‚Üí Built before 1800
- Modern ‚Üí Built in or after 1800

**üìä Plot: Average Visitors Comparison**

```python
df["Era"] = df["Year_Built_Num"].apply(
    lambda x: "Ancient" if x < 1800 else "Modern"
)

avg_visitors_era = df.groupby("Era")["Annual_Visitors_Millions"].mean()

plt.figure(figsize=(6,5))
plt.bar(avg_visitors_era.index, avg_visitors_era.values)
plt.xlabel("Attraction Era")
plt.ylabel("Average Annual Visitors (Millions)")
plt.title("Ancient vs Modern Attractions: Visitor Comparison")
plt.tight_layout()
plt.show()
```

### **üåç 3Ô∏è‚É£ Regions Rich in Historical Landmarks**
**üîπ Purpose**

- Identify regions with older historical sites
- Cultural richness comparison

**üìä Plot: Average Year Built by Region**

```python
avg_year_region = df.groupby("Region")["Year_Built_Num"].mean().sort_values()

plt.figure(figsize=(8,5))
plt.bar(avg_year_region.index, avg_year_region.values)
plt.xlabel("Region")
plt.ylabel("Average Year Built")
plt.title("Historical Richness of Regions (Average Year Built)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
