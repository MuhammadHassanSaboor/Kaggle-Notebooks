# üìä **Dataset Overview**
This dataset contains information about the population and geographical details of 234 countries and territories. Below is a summary of the dataset's structure:

- **Total Entries**: 234
- **Total Columns**: 17

## üìã **Columns Information:**
1. **Rank**: Rank of the country/territory (integer)
2. **CCA3**: 3-letter country code (string)
3. **Country/Territory**: Name of the country or territory (string)
4. **Capital**: Capital city of the country (string)
5. **Continent**: Continent where the country is located (string)
6. **2022 Population**: Population in 2022 (integer)
7. **2020 Population**: Population in 2020 (integer)
8. **2015 Population**: Population in 2015 (integer)
9. **2010 Population**: Population in 2010 (integer)
10. **2000 Population**: Population in 2000 (integer)
11. **1990 Population**: Population in 1990 (integer)
12. **1980 Population**: Population in 1980 (integer)
13. **1970 Population**: Population in 1970 (integer)
14. **Area (km¬≤)**: Area of the country in square kilometers (integer)
15. **Density (per km¬≤)**: Population density per square kilometer (float)
16. **Growth Rate**: Population growth rate (float)
17. **World Population Percentage**: Percentage of the world population (float)

# üîç Exploratory Data Analysis (EDA)

# üåç **Population Trends and Growth**

```python
population_growth = df.melt(
    id_vars=['Country/Territory'], 
    value_vars=['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population'],
    var_name='Year', 
    value_name='Population'
)
population_growth['Year'] = population_growth['Year'].str.extract('(\d+)').astype(int)
fig1 = px.line(population_growth, 
               x='Year', 
               y='Population', 
               color='Country/Territory', 
               title='Population Growth Over Time', 
               template='plotly_dark')

fig1.update_layout(paper_bgcolor='black', 
                   plot_bgcolor='black')
fig1.show()
```

```python
data_sorted = df.sort_values(by='Growth Rate', ascending=False)
highest_growth = data_sorted.head(10)
lowest_growth = data_sorted.tail(10)

fig2 = px.bar(highest_growth, 
              x='Country/Territory', 
              y='Growth Rate', 
              title='Countries with Highest Growth Rates', 
              template='plotly_dark', 
              color='Growth Rate')

fig2.update_layout(paper_bgcolor='black', 
                   plot_bgcolor='black')
fig2.show()
```

```python
fig3 = px.bar(lowest_growth, 
              x='Country/Territory', 
              y='Growth Rate', 
              title='Countries with Lowest Growth Rates', 
              template='plotly_dark', 
              color='Growth Rate')

fig3.update_layout(paper_bgcolor='black', 
                   plot_bgcolor='black')
fig3.show()
```

```python
decade_diff = df[['Country/Territory', '1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2020 Population']]
decade_diff['1970-1980'] = decade_diff['1980 Population'] - decade_diff['1970 Population']
decade_diff['1980-1990'] = decade_diff['1990 Population'] - decade_diff['1980 Population']
decade_diff['1990-2000'] = decade_diff['2000 Population'] - decade_diff['1990 Population']
decade_diff['2000-2010'] = decade_diff['2010 Population'] - decade_diff['2000 Population']
decade_diff['2010-2020'] = decade_diff['2020 Population'] - decade_diff['2010 Population']
```

```python
fig4 = px.box(
    decade_diff.melt(id_vars=['Country/Territory'], 
                     value_vars=['1970-1980', '1980-1990', '1990-2000', '2000-2010', '2010-2020'], 
                     var_name='Decade', 
                     value_name='Population Change'),
    x='Decade', 
    y='Population Change', 
    title='Decade-Wise Population Changes', 
    template='plotly_dark',
    color='Decade'
)
fig4.update_layout(paper_bgcolor='black', 
                   plot_bgcolor='black')
fig4.show()
```

```python
data_sorted_by_pop = df.sort_values(by='2022 Population', ascending=False)

top_10 = data_sorted_by_pop.head(10)
bottom_10 = data_sorted_by_pop.tail(10)

fig5 = px.bar(top_10, 
              x='Country/Territory', 
              y='2022 Population', 
              title='Top 10 Most Populous Countries (2022)', 
              template='plotly_dark', 
              color='2022 Population')

fig5.update_layout(paper_bgcolor='black', 
                   plot_bgcolor='black')
fig5.show()
```

```python
fig6 = px.bar(bottom_10, 
              x='Country/Territory', 
              y='2022 Population', 
              title='Top 10 Least Populous Countries (2022)', 
              template='plotly_dark', 
              color='2022 Population')

fig6.update_layout(paper_bgcolor='black', 
                   plot_bgcolor='black')
fig6.show()
```

# üåç **Geographical Insights**

```python
pie_2022 = px.pie(df, names='Continent', values='2022 Population', 
                  title='Population by Continent (2022)', 
                  template='plotly_dark', 
                  color_discrete_sequence=px.colors.sequential.Viridis)

pie_2020 = px.pie(df, names='Continent', values='2020 Population', 
                  title='Population by Continent (2020)', 
                  template='plotly_dark', 
                  color_discrete_sequence=px.colors.sequential.Viridis)

pie_2015 = px.pie(df, names='Continent', values='2015 Population', 
                  title='Population by Continent (2015)', 
                  template='plotly_dark', 
                  color_discrete_sequence=px.colors.sequential.Viridis)

pie_2010 = px.pie(df, names='Continent', values='2010 Population', 
                  title='Population by Continent (2010)', 
                  template='plotly_dark', 
                  color_discrete_sequence=px.colors.sequential.Viridis)

fig = make_subplots(rows=2, cols=2, 
                    subplot_titles=["2022", "2020", "2015", "2010"], 
                    specs=[[{'type': 'domain'}, {'type': 'domain'}],
                           [{'type': 'domain'}, {'type': 'domain'}]])

fig.add_trace(go.Pie(labels=pie_2022.data[0]['labels'], 
                     values=pie_2022.data[0]['values'], 
                     marker=dict(colors=px.colors.sequential.Viridis)), 
              row=1, col=1)

fig.add_trace(go.Pie(labels=pie_2020.data[0]['labels'], 
                     values=pie_2020.data[0]['values'], 
                     marker=dict(colors=px.colors.sequential.Viridis)), 
              row=1, col=2)

fig.add_trace(go.Pie(labels=pie_2015.data[0]['labels'], 
                     values=pie_2015.data[0]['values'], 
                     marker=dict(colors=px.colors.sequential.Viridis)), 
              row=2, col=1)

fig.add_trace(go.Pie(labels=pie_2010.data[0]['labels'], 
                     values=pie_2010.data[0]['values'], 
                     marker=dict(colors=px.colors.sequential.Viridis)), 
              row=2, col=2)

# Update layout
fig.update_layout(title_text="Population by Continent (2010-2022)", 
                  title_font_color='white', 
                  paper_bgcolor='black', 
                  font_color='white')

# Show the figure
fig.show()
```

```python
fig = make_subplots(rows=2, cols=2, 
                    specs=[[{'type': 'domain'}, {'type': 'domain'}],
                           [{'type': 'domain'}, {'type': 'domain'}]],
                    subplot_titles=("1970", "1980","1990", "2000"))

fig.add_trace(go.Pie(labels=df['Continent'], values=df['1970 Population'], 
                     marker=dict(colors=px.colors.sequential.Viridis), 
                     title="1970"), 
              row=1, col=1)

fig.add_trace(go.Pie(labels=df['Continent'], values=df['1980 Population'], 
                     marker=dict(colors=px.colors.sequential.Viridis), 
                     title="1980"), 
              row=1, col=2)

fig.add_trace(go.Pie(labels=df['Continent'], values=df['1990 Population'], 
                     marker=dict(colors=px.colors.sequential.Viridis), 
                     title="1990"), 
              row=2, col=1)

fig.add_trace(go.Pie(labels=df['Continent'], values=df['2000 Population'], 
                     marker=dict(colors=px.colors.sequential.Viridis), 
                     title="2000"), 
              row=2, col=2)

fig.update_layout(template='plotly_dark', 
                  title_font_color='white', 
                  paper_bgcolor='black', 
                  font_color='white', 
                  showlegend=False, 
                  title_text="Population by Continent Over Decades")

fig.show()
```

```python
fig2 = px.bar(df, 
              x='Continent', 
              y='Density (per km¬≤)', 
              title='Population Density by Continent', 
              template='plotly_dark', 
              color='Density (per km¬≤)', 
              color_continuous_scale=px.colors.sequential.Viridis)

fig2.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig2.show()
```

```python
fig3 = px.bar(df, 
              x='Continent', 
              y='World Population Percentage', 
              title='World Population Percentage by Continent (2022)', 
              template='plotly_dark', 
              color='World Population Percentage', 
              color_continuous_scale=px.colors.sequential.Viridis)

fig3.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig3.show()
```

# ‚ö†Ô∏è **Outliers and Distributions**

```python
df_outliers = df[['Country/Territory','Area (km¬≤)','Density (per km¬≤)','Growth Rate']]
```

```python
fig1 = px.box(df_outliers, 
              y='Growth Rate',
              title='Distribution and Outliers of Growth Rates',
              template='plotly_dark')

fig1.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig1.show()
```

```python
fig2 = px.histogram(df_outliers, 
                    x='Growth Rate', 
                    title='Distribution of Growth Rates',
                    template='plotly_dark', 
                    nbins=10, 
                    color_discrete_sequence=px.colors.sequential.Viridis)

fig2.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig2.show()
```

```python
# Density Clusters
X = df_outliers[['Area (km¬≤)', 'Density (per km¬≤)']]
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
df_outliers['Cluster'] = kmeans.labels_
```

```python
fig3 = px.scatter(df_outliers, 
                  x='Area (km¬≤)', 
                  y='Density (per km¬≤)', 
                  color='Cluster', 
                  title='Density Clusters',
                  template='plotly_dark',
                  color_discrete_sequence=px.colors.sequential.Viridis)

fig3.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig3.show()
```

# ‚è∞ **Historical Comparisons**

```python
df_historical = pd.DataFrame(df[['Country/Territory','1970 Population','2022 Population']])
df_historical['Absolute Change'] = df_historical['2022 Population'] - df_historical['1970 Population']
df_historical['Percentage Change'] = (df_historical['Absolute Change'] / df_historical['1970 Population']) * 100
```

```python
fig1 = px.bar(df_historical, 
              x='Country/Territory', 
              y='Absolute Change', 
              title='Top Gainers and Losers (Absolute Change 1970-2022)', 
              template='plotly_dark', 
              color='Absolute Change', 
              color_continuous_scale=px.colors.sequential.Viridis)

fig1.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig1.show()
```

```python
fig2 = px.line(df_historical.melt(id_vars=['Country/Territory'], 
                                value_vars=['1970 Population', '2022 Population'], 
                                var_name='Year', 
                                value_name='Population'), 
                                x='Year', 
                                y='Population', 
                                color='Country/Territory', 
                                title='Population Shifts (1970-2022)', 
                                template='plotly_dark', 
                                color_discrete_sequence=px.colors.qualitative.Vivid)

fig2.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig2.show()
```

# ‚öñÔ∏è **Comparisons by Size and Scale**

```python
df_size_scale = pd.DataFrame(df[['Country/Territory','Area (km¬≤)','Density (per km¬≤)','Growth Rate','2022 Population']])
df_size_scale['Size Category'] = ['Small' if x < 5000 else 'Large' for x in df_size_scale['Area (km¬≤)']]
```

```python
fig1 = px.box(df_size_scale, 
              x='Size Category', 
              y='Density (per km¬≤)', 
              title='Density Comparison: Small vs. Large Countries',
              template='plotly_dark', 
              color='Size Category')

fig1.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig1.show()
```

```python
fig2 = px.scatter(df_size_scale, 
                  x='Density (per km¬≤)', 
                  y='Growth Rate', 
                  size='2022 Population', 
                  color='Size Category', 
                  title='Urbanization Proxy: Density vs. Growth Rate',
                  template='plotly_dark',
                  color_discrete_sequence=px.colors.qualitative.Vivid)

fig2.update_layout(title_font_color='white', 
                   paper_bgcolor='black', 
                   font_color='white')
fig2.show()
```
